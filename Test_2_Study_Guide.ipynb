**Background Needed**

The Klenow Fragment is a large piece of DNA polymerase-1 taken from *Escherichia coli*. Unlike most poly-1 segments that have exonuclease ('proofreading') and polymerase activity in the 5'-3' and 3'-5' directions, the Klenow Fragment lacks activity on the 5'-3' direction (loss of exonuclease acitivity, DNA poly-1 has no polymerase activity in that direction regardless). The Klenow Fragment was discovered by Han Klenow & colleagues after treating *E. Coli* poly-1 with the enzyme *subtilisin* (a protease) to proteolytically cleave the polymerase. They initially started this experiment to probe polymerase-1 to better understand the structure and found that it had two seperate domains that held different functions. Ultimately, the Klenow Fragment is an incredibly useful protein as biochemists are more easily able to study polymerase acitivity, kinetics, etc. without having to consider unnecessary activity (i.e. exonuclease); one less factor to deal with in experimental design.

**Dahlberg & Benkovic's Kinetic Study of The Klenow Fragment**

The Klenow Fragment (KF) has been intensely researched for many years by the time this paper was written; it's structure, stereochemical properties, and kinetic actively had been studied extensively. The 'current' proposed kinetic mechanism (*Kuchta et al.*) in which a nucleotide is correctly incorporated was proposed to be the ordered binding of DNA then dNTP followed by a show nonchemical reaction event that limits the rate for a single incorporation event. After, phosphodiester bond formation occurs and is followed by PP<sub> i </sub> (inorganic phosphate) release. KF would then partition between more nucleotide incorporation or dissociation from DNA.

At the time of this paper, the kinetic mechanism for incorrect nucleotide addition was found by *Kutcha et al.* and was found to proceed via 3 distinct stages:
1. Nucleotide binding and chemical catalysis
2. Conformational change step that occurs post-catalysis but pre-PP<sub>i</sub> release
3. The partitioning (division) of the mismatched primer terminus between addition of a correct nucleotide and it exonucleolytic removal

Dalhberg & Benkovic state that although the conformation change in the second step has been well identified, they've found evidence indicating that this step is important in the pathway for the *correct* incorporation as well:

"In this report we describe evidence for the occurrence of this step through combination of pre-steady-state pyrophosphorolysis and pyrophosphate-exchange measurements and pulse-chase experiments, which serve to locate this step and to identify the ternary intermediates separated by it."

*Quench Comparison Studies*
>Dahlberg et al. ran rapid quench experiments of the E-DNA<sub>13</sub> + dATP<sup>*</sup>, Mg<sup>2+</sup> in EDTA, HCl (acid), and NaOH (base). They found that in base, the results were indicative of poor quenching since total incorporation was observed at all time points (failed quench). EDTA did quench and lead to 100% product formation. This is interesting because EDTA strips off the metal cofactor (Mg<sup>2+</sup>) off, which normally would stop it. The fact that 100% product was still formed tells us that, even for a very short period, any intermediate formed after the first step is fully committed and will progress to the next step in the reaction. HCl was the stand out quench as it's burst amplitude was 20% shorter than EDTA, but what does that mean? Dahlberg et al. state that the pre-steady state kinetics are identical between HCl and EDTA, but the _final burst amplitude_ showed that HCl quenched at 80% product formation. This tells us that:
>1. HCl can trap the intermediate
>2. As long as the first conformational change has occurred, the reaction can proceed without the metal cofactor
>3. A kinetically significant intermediate accumulates during correct incorporation

## *Nicotinamides*

**Background Needed**

Nicotinamide coenzymes (Pyridine nucleotides) such as NAD<sup>+</sup> and NADP<sup>+</sup> are cofactors used to carry out redox reactions in biological systems. Nicotinamides are only capable of reducing/oxidizing by 1 e<sup>-</sup> at a time (1 proton) and can be paired with othe redox cofactors like Flavins. Nicotinamides are bound to an enzyme's active site via noncovalent interactions. An enzyme without it's cofactor is called an _apoenzyme_; with its cofactor its a _holoenzyme_. Cofactors can be fairly easily removed from enzymes via gel filtration or dialysis. Cofactors like nicotinamides are no biosynthesized and are derived from our diets; Pellagra is a nutritional disorder thats defined by a lack of coenzymes in the diet.

**Alcohol Dehydrogenase**

Alcohol dehydrogenase (ADH) are enzymes that are responsible for catalyzing the reversible oxidation of alcohols to the corresponding aldehydes. This mechanism suggests a hydride mechanism, which is observed in (most) enzymatic reactions. Although, they can proceed through radical reactions but that is rare in organic contexts. When reducing an aldehyde with ADH, a new chiral center is typical to form. ADHs are very selective and require very specific geometry and conformation for the reaction to carry out.

**Amino Acid Dehydrogenase**

Very similar to the ADH mechanism, but uses a lysine and glutamate residue to complete proton transfers. Glutamate will participate in a hydride transfer with the amine of the acid forming a Shiff base, which then takes a proton from NADH (NADH -> NAD<sup>+</sup>) oxidizing the cofactor. Lysine then deprotonates water, which then attacks the azomethine carbon of the Shiff base. From there, NH2 will become protonated and act as a leaving group, with the new -OH bond (deprot water attack) forming a double bond at the central carbon and forming the final product.

## *General Experimental Techniques*

_Radical Clock_

We've gone over multiple methods that can be used to find a covalent intermediate (burst kinetics, radiolabeling, etc.) but how can you determine if a enzyme uses a radical mechanism? That's where the radical clock comes in. A radical clock in a mechanistic probe that can be used to detect and time radical intermediates in a reaction. A radical clock relys on the molecules ability to carrying out a unimolecular radical rearrangement (hydrogen _atom_ shift); a radical clock will not work otherwise. The main idea of a radical clock is to do a differential rate analysis between the altered molecules and the wild type. For example if I was studying a P-450 monooxygenase hydroxylation of the C17 on a steroid, I'd want to:
1. Confirm the steroid's ability to have a cyclopropane formation occur on the C17-branch
2. Find the rate of the desired step using the wild type substrate
3. If (1) = true, find the rate using the mutant susbtrate

After that, I can compare both the products and rate to determine whether or not a radical mechanism is used by the enzyme.

_Rapid Quench-Flow_

Rapid quench-flow tests are very similar to stop flow, but at a higher precision and different applicaiton. Rapid quench-flow is a pre-steady state technique that allows you to see the chemical events of a reaction on the millisecond timescale. The basic setup involves 3 syringes at 3 different points: a syringe containing enzymes + cofactors, parallel to a syringe containing substrate, where the lines then combine and meet in a central tube to mix where the third syringe filled with quenching agent is located. The quenching agent instantly stops (quenches) the reaction, preserving the concentration of all intermediates, substrates, and products in the solution at that point in time. The idea is that you're taking chemical snapshots, where you can then measure the concentration of intermediate to product. Rapid quench-flow is powerful because you're taking these measurements pre-steady state, allowing you to determine rate-limiting conformational changes, intermediate accumulation, burst kinetics and more.

_Pulse Chase_

Pulse chase is a method that exploits a complexes commitment to catalysis and physical release of product. You use pulse chase if you're asking: "Once an enzyme binds to substrate, has it truly committed to completing the reaction?". The purpose of the pulse chase method is to determine the partitioning ratio, k<sub>off</sub> & k<sub>forward</sub>. This is valuable for finding the rate limiting step in a pathway. The set up for this method is to prepare labeled substrate and begin the reaction with enzyme, E + S* -> E · S* -> E · P* -> E + P* and let it run for a short but defined amount of time, _t_. Once _t_ elapses, flood the reaction with xs. unlabeled substrate. After a set amount of time, measure the [P*]. The goal is to complete differential rate analysis between the k<sub>off</sub> and k<sub>forward</sub> to determine which is the rate limiting step. If k<sub>off</sub> > k<sub>forward</sub> => the enzyme is not 'very committed' & the chemistry step is rate limiting (note: when I say rate limiting I am stating that the step is limiting to the reaction rate; 'fuel is the limiting variable to distance'). If k<sub>off</sub> < k<sub>forward</sub> => the substrate binding step is rate limiting. If k<sub>off</sub> = k<sub>forward</sub> => uhhhhh indeterminate the rate of each step is equal to one another (?).
