## ***Substitutions***

There are multiple sub-topics in substitutions and its the least-recent thingy we covered so theres gonna be a bit more content in this section. To begin, (nucleophilic) substitutions are when one group attached to the carbon gets replaced, substituted, with another group. The attacking group is called the nucleophile, while the ejected group is called the leaving group. For nucleophilic subsitions, there are 3 primary 'players' we must consider:
- the nucleophile (attacking group)
- the electrophile (point of reaction)
- the leaving group (ejected group)

There are generally 2 types of substitution reactions, S<sub>n</sub>1 & S<sub>n</sub>2 reactions. S<sub>n</sub>1 reactions are stepwise and unimolecular (2 step). The LG (leaving group) departs first, generating a discrete carbocation intermediate, only then does the nucleophile come in and attack the electrophilic region; the starting stereochemistry is conserved.

S<sub>n</sub>2 reactions are concerted and bimolecular (1 step). The LG gets ejected by the simultaneous backside attack of the nucleophile. The nucleophile is forced to attack the electrophile on the opposite (back) side of the LG causing the stereochemistry to invert.

For the following enzymes, we should ask the following questions:
1. What is the electrophile and why is it electrophilic?
2. What is the LG and how is the substrate stabilized after it's ejection?
3. What is the nucleophile and how does the enzyme position/activate it?
4. Is this an S<sub>n</sub>1 or S<sub>n</sub>2 reaction and what evidence do we have to support this?

### *Terpenes*

There are 3 terpenes centered substituion enzymes/reactions that we need to cover:
- IPP Isomerase
- Geranyl Diphosphate Synthase
- Farnesyl-DP Synthase

IPP Isomerase is by far the simplest, setting the template for how we should expect these reactions to go down: **Ionization of an allylic diphosphate to generate a carbocation, followed by nucleophilic capture by an alkene π bond**. Isopentyl Diphosphate (IPP) is initially generated by 2 acetyl-CoA molecules reacting together with CO<sub>2</sub> as a byproduct. IPP then undergoes isomerization where a singular protanted base group, protonates (lock in bro) the alkene of IPP forming a tertiary carbocation intermediate. The secondary α-carbon of the carbocation then gets deprotonated by the base group, creating an alkene between the secondary and tertiary carbon; product is Dimethyl Diphosphate (Di-PP). It's a simple 2 step reaction (S<sub>n</sub>1) and it's the basis for how the following reactions will go down

Geranyl Diphosphate Synthase (GDS) might seem taunting at first but this too is a pretty simple polymerization reaction. Geranyl diphosphate is a steroid precursor so it was once a high interest enzyme. What's unique about GDS is that it uses a Mg<sup>2+</sup> ligand to stabilize the negative charges on the diphosphate; the polymerization wouldn't be able to proceed otherwise. The mechanism is identical to IPP Isomerase except the base used *is* isopentyl diphosphate. The reactions essentially works with 1 mole Di-PP reacting with 1 mole IPP, with the diphosphate of the Di-PP acting as an LG (-PP<sub>i</sub>). The exit of PP creates a carbocation on the terminal primary carbon, which then reacts with the IPP. The IPP molecule then forms a carbocation on the tertiary carbon (as we previously saw) and the secondary carbon gets deprotanted by the PP<sub>i</sub> group, reforming the alkene and finishing the Geranyl Disphosphate synthesis. The Farnesyl-DP Synthase proceeds in the same way, except the substrate is Geranyl-DP instead of Di-PP.

All this information is fantastic, but *how do we prove this*? The primary characteristic of the reactions that we should be looking for is the existance of a carbocation, the problem is F-DP Synthase is hyper selective and we're unfortunately not able to make many changes to the substrate or enzyme to analyze it. Researchers found that because flourine is fairly similar to hydrogen in size (high electronegativity -> compacts outer orbitals reducing size), they can insert it into the substrate to replace certain protons to probe the prescence of a carbocation. If the mechanism proceeded via a carbocation, the prescence of a adjacent fluorine would decrease k<sub>cat</sub> by destabilizing the carbocation. A radical or anion mechanism would not be impacted. After the running the reactions with the modified substrates, they did find that fluorine impacted the k<sub>cat</sub>, thus proving the Farnesyl-DP proceeded via a carbocation intermediate.
